```python
df_x= df.groupby('Customer ID').agg({'total_price': lambda x: x.sum(),
                                        'InvoiceDate': lambda x: (today - x.max()).days})
# Ä°lki monetory, ikincisi recency
```

```python
df_x
```

```python
df_y = df.groupby(['Customer ID','Invoice']).agg({'total_price': lambda x: x.sum()})
```

```python
df_y
```

```python
df_z = df_y.groupby('Customer ID').agg({'total_price': lambda x: len(x)})
```

```python
df_z
```

# **RFM TABLE**

```python
rfm_table=pd.merge(df_x,df_z,on="Customer ID")
```

```python
rfm_table.rename(columns={"total_price_x": "Monetary","InvoiceDate":"Recency","total_price_y":"Frequency"}, inplace=True)
```

```python
rfm_table
```

```python
def FScore(x,p,d):
    if x <= d[p][0.20]:
        return 0
    elif x <= d[p][0.40]:
        return 1
    elif x <= d[p][0.60]: 
        return 2
    elif x <= d[p][0.80]:
        return 3
    else:
        return 4
```

# Frequency puanÄ± bulma

```python
quantiles = rfm_table.quantile(q=[0.20,0.40,0.60,0.80])
quantiles = quantiles.to_dict()
rfm_table['Freq_Tile'] = rfm_table['Frequency'].apply(FScore, args=('Frequency',quantiles))
```

```python
rfm_table
```

# Recency puanÄ± bulma

```python
rfm_table = rfm_table.sort_values('Recency',ascending=True)
```

```python
rfm_table
```

```python
rfm_table['Rec_Tile'] = pd.qcut(rfm_table['Recency'],5,labels=[4,3,2,1,0])
```

```python
rfm_table
```

# Monetary skoru bulma

```python
rfm_table.sort_values('Monetary',ascending=False)
rfm_table['Mone_Tile'] = pd.qcut(rfm_table['Monetary'],5,labels=False)
```

```python
rfm_table
```

# RFM skoru bulma

```python
rfm_table.info()
```

```python
rfm_table["Rec_Tile"] = pd.to_numeric(rfm_table["Rec_Tile"])
```

```python
rfm_table['RFM Score'] = rfm_table['Rec_Tile'].map(str) + rfm_table['Freq_Tile'].map(str) + rfm_table['Mone_Tile'].map(str)
```

```python
rfm_table
```

# inceleme

```python
rfm_table.groupby('RFM Score').agg({
'Recency': ['mean','min','max','count'],
'Frequency': ['mean','min','max','count'],
'Monetary': ['mean','min','max','count'] }).round(1)
```

```python
rfm_table['RF Score'] = rfm_table['Rec_Tile'].map(str) + rfm_table['Freq_Tile'].map(str)
```

```python
rfm_table
```

```python
seg_map  = { r'[0-1][0-1]' : 'hibernating',
         r'[0-1][2-3]' : 'at_risk',
         r'[0-1]4' : 'cant_loose',
         r'2[0-1]' : 'about_to_sleep',
         r'22' : 'need_attention',
         r'[2-3][3-4]' : 'loyal_customers',
         r'30' : 'promising',
         r'40' : 'new_customers',
         r'[3-4][1-2]' : 'potantial_loyalist',
         r'4[3-4]' : 'champions' }

rfm_table['RF_SEGMENTS'] = rfm_table['RF Score'].replace( seg_map, regex=True )
```
